# Data Collection and Preparation

```python
import numpy as np
import pandas as pd
from datetime import date, time, datetime, timedelta
from matplotlib import pyplot
import matplotlib.pyplot as plt
```

## Weather Data

The weather data used is publicly accessible from the weather.com api and contains the following variables. <br>

<b>date_time</b>: Day and time of measurement <br>
<b>max_temp</b>: Maximum temperature recorded in °C in the respective period<br>
<b>min_temp</b>: Minimum temperature recorded in °C in the respective period<br>
<b>precip</b>: Binary indicator for precipitation (1=yes, 0=no) in the respective period<br>

```python
weather=pd.read_csv('../data/LosAngeles/weather_hourly_la.csv')
weather['date_time'] = pd.to_datetime(weather['date_time'])
weather = weather[(weather['date_time']>=datetime(year=2018,month=1,day=1)) &
                 (weather['date_time']<datetime(year=2019,month=1,day=1))]
weather
```

```python
weather.info()
```

The data set does not include null values.

```python
weather.describe()
```

The minumum and maximum values for max_temp and min_temp are realistic for the city of Los Angeles, California.<br>
There does not appear to be any outliers in the weather data.

### Delete duplicates

```python
weather[weather.duplicated(keep=False)]
```

The data set contains 214 rows with identical values for all four variables in two or more rows. We will delete these entries and only keep the first occurence of dupicated rows. 

```python
weather.drop_duplicates(keep = 'first', inplace = True) 
weather
```

```python
weather[weather.duplicated(subset=['date_time'], keep=False)]
```

For entries of the same hour with different values for max_temp and min_temp, we will use the mean value of the variable. 

```python
weather_clean = weather.groupby('date_time', as_index = False).mean()
weather_clean
```

```python
print(weather_clean.precip.unique())
weather_clean.loc[weather_clean['precip']==0.5,'precip'] = np.NaN
weather_clean[weather_clean['precip'].isnull()]
```

The feature precip is a binary indicator. We will use the last valid value of precipitation for different values of precip and the same value of date_time. 

```python
weather_clean.fillna(method='ffill', inplace = True)
weather_clean.info()
```

### Fill missing hours

The data set contains 8305 entries. Since the year 2018 had 8760 hours (365*24), we will add the missing hours to the time series data. 

```python
# create dataframe with missing hours
complete_date_time = pd.date_range(start = '2018-01-01 00:00:00', end = '2018-12-31 23:00:00', freq='H').tolist()
missing_date_time = set(complete_date_time) - set(weather_clean['date_time'])
missing_df = pd.DataFrame(missing_date_time, columns=['date_time'])

# add missing hours to weather data
complete_weather = pd.merge(weather_clean,missing_df,on='date_time', how='outer')
complete_weather.sort_values(by=["date_time"], inplace = True)
complete_weather.set_index('date_time', inplace=True)
complete_weather
```

The data set contains no more than eight consecutive null values. We assume a linear and equally spaced change of minimum and maximum temperature for the missing time intervals and therefore use the interpolate method 'linear' to impute the null values. 

```python
fig,ax = plt.subplots(figsize=(20,10), dpi= 80) 
axis_font = {'size':'20'}
label_font = {'labelsize':'12'}
dic={'fontsize': '24'}

ax.plot(complete_weather['max_temp'], color = 'tab:blue')
ax.set_title("Weather Data before Imputation", dic)
ax.set_ylabel('maximum temperature',**axis_font)
ax.set_xlabel('date',**axis_font)
ax.tick_params(axis='both', **label_font)
plt.show()

complete_weather = complete_weather.assign(max_temp=complete_weather.max_temp.interpolate(method='linear', limit = 8))

fig,ax = plt.subplots(figsize=(20,10), dpi= 80) 

ax.plot(complete_weather['max_temp'], color = 'tab:blue')
ax.set_title("Weather Data after Imputation", dic)
ax.set_ylabel('maximum temperature',**axis_font)
ax.set_xlabel('date',**axis_font)
ax.tick_params(axis='both', **label_font)
plt.show()
```

```python
fig,ax = plt.subplots(figsize=(20,10), dpi= 80) 

ax.plot(complete_weather['min_temp'], color = 'tab:blue')
ax.set_title("Weather Data before Imputation", dic)
ax.set_ylabel('minimum temperature',**axis_font)
ax.set_xlabel('date',**axis_font)
ax.tick_params(axis='both', **label_font)
plt.show()

complete_weather = complete_weather.assign(min_temp=complete_weather.min_temp.interpolate(method='linear', limit = 8))

fig,ax = plt.subplots(figsize=(20,10), dpi= 80) 

ax.plot(complete_weather['min_temp'], color = 'tab:blue')
ax.set_title("Weather Data after Imputation", dic)
ax.set_ylabel('minimum temperature',**axis_font)
ax.set_xlabel('date',**axis_font)
ax.tick_params(axis='both', **label_font)
plt.show()
```

Precip is a binary indicator for precipitation. We assume whether precipitation recorded in the missing hours is the same as in the preceding hour. We therefore replace the missing values with the last valid entry.

```python
fig,ax = plt.subplots(figsize=(20,10), dpi= 80) 

ax.plot(complete_weather['precip'], color = 'tab:blue')
ax.set_title("Weather Data before Imputation", dic)
ax.set_ylabel('preciption',**axis_font)
ax.set_xlabel('date',**axis_font)
ax.tick_params(axis='both', **label_font)
ax.set_yticks([0,1])
plt.ylim(0, 1.1)
plt.show()

complete_weather = complete_weather.assign(precip=complete_weather.precip.fillna(method = 'ffill'))

fig,ax = plt.subplots(figsize=(20,10), dpi= 80) 

ax.plot(complete_weather['precip'], color = 'tab:blue')
ax.set_title("Weather Data after Imputation", dic)
ax.set_ylabel('preciption',**axis_font)
ax.set_xlabel('date',**axis_font)
ax.tick_params(axis='both', **label_font)
ax.set_yticks([0,1])
plt.ylim(0, 1.1)
plt.show()
```

```python
complete_weather.describe()
```

```python
complete_weather.reset_index(inplace=True)
complete_weather.info()
```

The resulting data set includes all 8760 hours of 2018 and no null values.

### Exporting weather data

```python
complete_weather.to_csv(r'../data/weather.csv', index = False)
complete_weather
```
