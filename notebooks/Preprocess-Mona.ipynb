```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_LA = pd.read_csv("la_2018.csv")
df_LA.drop_duplicates()
```

```python
df_LA.info()
```

```python
df_LA['start_time'] = pd.to_datetime(df_LA['start_time'])
df_LA['end_time'] = pd.to_datetime(df_LA['end_time'])
df_LA.info()
```

```python
def calculate_trip_time(df):
    # calculate the total trip duration in hours
    df['trip_duration'] = df.apply(lambda x: x['end_time'] - x['start_time'], axis=1) / pd.Timedelta(hours=1)
    return df
```

```python
calculate_trip_time(df_LA)
```
